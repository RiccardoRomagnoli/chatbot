```
import pandas as pd
import numpy as np
import spacy
from sklearn.feature_extraction import stop_words
```

# With Lemmatization

```
def lemmatize_text(input_list):
    spacy_model = spacy.load('en')
    lemmatized_descriptions = []
    for desc in input_list:
        current_desc = []
        doc = spacy_model(desc)
        for token in doc:
            current_desc.append(token.lemma_)
        lemmatized_descriptions.append(" ".join(current_desc))
    return lemmatized_descriptions
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/NLP Project/data.csv", encoding = "ISO-8859-1")


from sklearn.feature_extraction.text import TfidfVectorizer
df = df[df['Type'] != 0]
df = df.reset_index(drop=True)
corpus = list(df['user1'])

for i,item in enumerate(corpus):
    corpus[i]=corpus[i].lower()
lemmatized_corpus = lemmatize_text(corpus)
# vectorizer = TfidfVectorizer(stop_words = list(stop_words.ENGLISH_STOP_WORDS) + ['a', 'the','python', 'should', 'want', 'use', 'pron'], ngram_range=(1,1))
vectorizer = TfidfVectorizer(stop_words = ['a', 'the','python', 'should', 'want', 'use', 'pron'], ngram_range=(1,1))
X = vectorizer.fit_transform(lemmatized_corpus)

from scipy.sparse import csr_matrix
k=csr_matrix(X).toarray()
max(k[0])
```

```
correct = 0
correct_2 = 0

from scipy import spatial
df1=pd.read_csv("/content/drive/My Drive/Colab Notebooks/NLP Project/data_modified.csv", encoding="ISO-8859-1")
df1=df1[df1['modified']==1]
df1 = df1.reset_index(drop=True)
lemmatized_qs = lemmatize_text(list(df1['user1']))
for i,qs in enumerate(lemmatized_qs):
    v=vectorizer.transform([qs.lower()])
    scores=[]
    for item in k:
        scores.append(1- spatial.distance.cosine(item,csr_matrix(v).toarray()))
    scores=np.array(scores)
    print('QS: '+qs)
    print('Ans: ')
    for item in scores.argsort()[-3:][::-1]:
        if scores[item] > 0.45:
            print(df['user2'][item])
            print(scores[item])
            if df['user2'][item] == df1['user2'][i]:
                correct += 1
            else:
                print('wrong')
            break
        else:
            print('Sorry i cannot answer this question yet :)')
            if df1['user2'][i] == 'Sorry i cannot answer this question yet :)':
                correct += 1
                correct_2 += 1
            else:
                print('wrong')
            break
    print()
    print(correct_2)
    print(correct)
    print(len(df1))
```

### Using array for k

```
k = []
vectorizers = []
dffs = []
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/NLP Project/data.csv", encoding="ISO-8859-1")
for cat in range(2, 7):
    if cat == 2:  #represents category 0
      vectorizer = TfidfVectorizer(stop_words=None, ngram_range=(1, 1))
      vectorizers.append(vectorizer)
      dff = df[df['Type'] == 0]
    else:
      vectorizer = TfidfVectorizer(stop_words=['a', 'the', 'python', 'should', 'want', 'use', 'pron'],
                                    ngram_range=(1, 1))
      vectorizers.append(vectorizer)
      dff = df[df['Type'] == cat]
    dff = dff.reset_index(drop=True)
    dffs.append(dff)
    corpus = list(dff['user1'])
    lemmatized_corpus = lemmatize_text(corpus)
    X = vectorizer.fit_transform(lemmatized_corpus)
    k.append(csr_matrix(X).toarray())
print(k[0])
print(len(k))
print(vectorizers[0])
print(len(vectorizers))
```

```
question = "What's up"
threshold = [0.8, 0.5, 0.55, 0.55, 0.5]
cat = 3
c = 0 if cat == 0 else cat-2
print(c)
lemmatized_qs = lemmatize_text([question])
for i, qs in enumerate(lemmatized_qs):
    v = vectorizers[c].transform([qs.lower()])
    scores = []
    for item in k[c]:
        scores.append(1 - spatial.distance.cosine(item, csr_matrix(v).toarray()))
    scores = np.array(scores)
    index = scores.argsort()[-3:][::-1][0]
    if scores[index] > threshold[c]:
        print(index)
        print(dffs[c]['user2'][index])
    else:
        print('Sorry i cannot answer this question yet :)')
```
